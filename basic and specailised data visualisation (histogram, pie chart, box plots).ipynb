## Two Plotting Styles in Matplotlib

### Option 1: Scripting Layer (Procedural)
Uses `plt.title()`, `plt.ylabel()`, `plt.xlabel()`, etc.

```python
df.plot(kind='area')
plt.title('Title')
plt.ylabel('Y Label')
plt.xlabel('X Label')
plt.show()
```

### Option 2: Artist Layer (Object-Oriented) — Preferred
Stores the plot in a variable and uses `ax.set_title()`, `ax.set_ylabel()`, `ax.set_xlabel()`.

```python
ax = df.plot(kind='area', alpha=0.35, figsize=(20, 10))
ax.set_title('Title')
ax.set_ylabel('Y Label')
ax.set_xlabel('X Label')
```

> **Note:** `df.plot(kind='area')` is equivalent to `df.plot.area()`

---
## Pandas `df.plot()` — Available Plot Types

| `kind=` | Plot Type | Equivalent Method |
|---------|-----------|-------------------|
| `'line'` | Line plot | `df.plot.line()` |
| `'area'` | Area plot | `df.plot.area()` |
| `'hist'` | Histogram | `df.plot.hist()` |
| `'bar'` | Vertical bar chart | `df.plot.bar()` |
| `'barh'` | Horizontal bar chart | `df.plot.barh()` |
| `'pie'` | Pie chart | `df.plot.pie()` |
| `'box'` | Box plot | `df.plot.box()` |
| `'scatter'` | Scatter plot | `df.plot.scatter()` |
| `'kde'` / `'density'` | Kernel density estimate | `df.plot.kde()` |
| `'hexbin'` | Hexagonal bin plot | `df.plot.hexbin()` |

### Common Parameters
- `figsize=(width, height)` — Set figure dimensions
- `alpha` — Transparency (0 = transparent, 1 = opaque)
- `color` — Color(s) for the plot
- `title` — Plot title
- `legend` — Show/hide legend
- `stacked` — Stack series (for area, bar, hist)
- `fontsize` — Tick label font size
- `grid` — Show gridlines (`True`/`False`)
- `rot` — Rotation angle for tick labels
- `ax` — Matplotlib axes object to plot on (for subplots)

## Sample Data for Demonstrations

The cells below create small sample datasets so every plot runs out of the box.

````python
# ── sample data used throughout the notebook ──

np.random.seed(42)
years = list(range(1980, 2014))

# Top-5 countries (simulated immigration counts)
df_top5 = pd.DataFrame({
    'India':       np.random.randint(5000, 40000, len(years)),
    'China':       np.random.randint(4000, 38000, len(years)),
    'UK':          np.random.randint(2000, 15000, len(years)),
    'Philippines': np.random.randint(3000, 30000, len(years)),
    'Pakistan':    np.random.randint(2000, 12000, len(years)),
}, index=years)

# Continent totals (for pie chart)
df_continents = pd.DataFrame({
    'Total': [15000, 250000, 82000, 38000, 70000, 5000]
}, index=['Africa', 'Asia', 'Europe', 'Latin America', 'Northern America', 'Oceania'])

# Single-country time series (for box plot)
df_japan = pd.DataFrame({'Japan': np.random.randint(100, 800, len(years))}, index=years)

# Scatter / bubble data
df_tot = pd.DataFrame({'year': years,
                        'total': np.cumsum(np.random.randint(500, 3000, len(years)))})

# Nordic countries (for multi-series histogram)
df_nordic = pd.DataFrame({
    'Denmark': np.random.randint(50, 600, len(years)),
    'Norway':  np.random.randint(50, 500, len(years)),
    'Sweden':  np.random.randint(50, 700, len(years)),
}, index=years)

# Two countries for bubble plot
df_bubble = pd.DataFrame({
    'Year':      years,
    'Brazil':    np.random.randint(1000, 15000, len(years)),
    'Argentina': np.random.randint(500, 8000, len(years)),
})

print('Sample data created ✓')
````

## 1. Area Plots

Area plots display cumulative trends across multiple categories over time. They are useful for showing composition changes.

**Key points:**
- Area plots are **stacked by default** (`stacked=True`)
- Each column must be either all positive or all negative values (NaN defaults to 0)
- Unstacked plots have a default transparency (alpha) of 0.5

**All Parameters:**
| Parameter | Description |
|-----------|-------------|
| `kind='area'` | Specifies area plot |
| `stacked=True/False` | Stacked (default) or overlapping |
| `alpha` | Transparency (0–1) |
| `figsize=(w, h)` | Figure dimensions |
| `color` / `colormap` | Colours for each series |
| `linewidth` / `lw` | Border line width |
| `linestyle` / `ls` | Border line style (`'-'`, `'--'`, `'-.'`, `':'`) |
| `legend` | Show legend |
| `fontsize` | Tick label size |
| `grid` | Show gridlines |
| `xlim` / `ylim` | Axis limits |

````python
# ── Stacked area plot (default) with full customisation ──

colours = ['#e6194B', '#3cb44b', '#4363d8', '#f58231', '#911eb4']

ax = df_top5.plot(
    kind='area',
    stacked=True,              # default — stacked
    alpha=0.65,                # transparency
    figsize=(14, 7),           # figure size
    color=colours,             # custom colour per series
    linewidth=1.5,             # border line width
    fontsize=12,               # tick label size
    grid=True,                 # show grid
)

ax.set_title('Immigration Trend of Top 5 Countries (Stacked)',
             fontsize=16, fontweight='bold', pad=15)
ax.set_ylabel('Number of Immigrants', fontsize=13)
ax.set_xlabel('Years', fontsize=13)
ax.legend(loc='upper left', fontsize=11, frameon=True,
          fancybox=True, shadow=True, borderpad=1)
ax.set_xlim(1980, 2013)

plt.tight_layout()
plt.show()
````

````python
# ── Unstacked area plot with transparency ──

ax = df_top5.plot(
    kind='area',
    stacked=False,             # overlapping instead of stacked
    alpha=0.30,                # lower alpha so overlap is visible
    figsize=(14, 7),
    colormap='Set2',           # use a named colourmap instead of list
    linewidth=2,
    linestyle='--',            # dashed border lines
    fontsize=12,
    grid=True,
)

ax.set_title('Immigration Trend of Top 5 Countries (Unstacked)',
             fontsize=16, fontweight='bold', fontstyle='italic', pad=15)
ax.set_ylabel('Number of Immigrants', fontsize=13)
ax.set_xlabel('Years', fontsize=13)
ax.legend(loc='upper right', fontsize=11, ncol=2,
          framealpha=0.9, edgecolor='black')

plt.tight_layout()
plt.show()
````

## 2. Histograms

A histogram represents the **frequency distribution** of a numeric dataset. It partitions the x-axis into **bins**, assigns each data point to a bin, and counts the number of data points per bin.

**All Parameters:**
| Parameter | Description |
|-----------|-------------|
| `kind='hist'` | Specifies histogram |
| `bins` | Number of intervals (default 10) |
| `xticks` | Custom bin-edge tick labels |
| `color` | Single colour or list of colours |
| `edgecolor` | Colour of bar borders |
| `linewidth` | Width of bar borders |
| `stacked` | Stack overlapping series |
| `alpha` | Transparency for overlapping |
| `orientation` | `'horizontal'` or `'vertical'` (default) |
| `rwidth` | Relative bar width (0–1), adds gaps between bars |
| `density` / `normed` | Normalise to show probability density |
| `cumulative` | Cumulative histogram |

````python
# ── Getting bin edges with NumPy ──
count, bin_edges = np.histogram(df_top5['India'])
print('Frequency counts :', count)
print('Bin edges        :', bin_edges)
````

````python
# ── Single histogram with maximum detail ──

count, bin_edges = np.histogram(df_top5['India'], bins=15)

ax = df_top5['India'].plot(
    kind='hist',
    bins=15,                       # number of bins
    figsize=(12, 6),
    color='steelblue',             # bar fill colour
    edgecolor='white',             # border colour between bars
    linewidth=1.2,                 # border width
    alpha=0.85,                    # slight transparency
    rwidth=0.9,                    # bar width (adds small gap)
    xticks=bin_edges,              # align ticks to bin edges
    fontsize=11,
    grid=True,
)

ax.set_title('Histogram of Indian Immigration (1980–2013)',
             fontsize=16, fontweight='bold', pad=15)
ax.set_ylabel('Number of Years (Frequency)', fontsize=13)
ax.set_xlabel('Number of Immigrants', fontsize=13)
ax.tick_params(axis='x', rotation=45)

# Add mean line
mean_val = df_top5['India'].mean()
ax.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean = {mean_val:,.0f}')
ax.legend(fontsize=12)

plt.tight_layout()
plt.show()
````

````python
# ── Multi-series histogram (overlapping, multiple countries) ──

ax = df_nordic.plot(
    kind='hist',
    bins=15,
    figsize=(12, 6),
    alpha=0.55,                              # transparency for overlap
    color=['coral', 'darkslateblue', 'mediumseagreen'],
    edgecolor='black',
    linewidth=0.8,
    stacked=False,                           # overlapping bars
    fontsize=11,
    grid=True,
)

ax.set_title('Immigration from Nordic Countries (1980–2013)',
             fontsize=16, fontweight='bold', pad=15)
ax.set_ylabel('Number of Years', fontsize=13)
ax.set_xlabel('Number of Immigrants', fontsize=13)
ax.legend(fontsize=12, frameon=True, shadow=True)

plt.tight_layout()
plt.show()
````

````python
# ── Stacked histogram variant ──

ax = df_nordic.plot(
    kind='hist',
    bins=12,
    figsize=(12, 6),
    alpha=0.80,
    color=['#e6194B', '#3cb44b', '#4363d8'],
    edgecolor='white',
    linewidth=1,
    stacked=True,                            # stacked bars
    fontsize=11,
    grid=True,
)

ax.set_title('Nordic Immigration — Stacked Histogram',
             fontsize=16, fontweight='bold', pad=15)
ax.set_ylabel('Cumulative Frequency', fontsize=13)
ax.set_xlabel('Number of Immigrants', fontsize=13)
ax.legend(fontsize=12)

plt.tight_layout()
plt.show()
````

## 3. Bar Charts

Bar charts are used for **categorical comparisons**, unlike histograms which show frequency distributions of continuous data.

**All Parameters:**
| Parameter | Description |
|-----------|-------------|
| `kind='bar'` | Vertical bar chart |
| `kind='barh'` | Horizontal bar chart |
| `color` | Single colour, list, or colourmap |
| `edgecolor` | Border colour of each bar |
| `linewidth` | Border line width |
| `width` | Bar width (0–1, default 0.5) |
| `stacked` | Stack grouped bars |
| `align` | `'center'` (default) or `'edge'` |
| `rot` | X-tick label rotation angle |
| `log` | Logarithmic y-axis |

````python
# ── Vertical bar chart with full customisation ──

# Use a subset for readability
df_bar = df_japan.tail(15)  # last 15 years

ax = df_bar.plot(
    kind='bar',
    figsize=(14, 6),
    color='teal',                  # bar fill colour
    edgecolor='black',             # bar border colour
    linewidth=0.8,                 # border width
    width=0.75,                    # bar width
    alpha=0.85,
    fontsize=11,
    grid=True,
    legend=False,
    rot=45,                        # rotate x-labels
)

ax.set_title('Japanese Immigrants to Canada (1999–2013)',
             fontsize=16, fontweight='bold', pad=15)
ax.set_ylabel('Number of Immigrants', fontsize=13)
ax.set_xlabel('Year', fontsize=13)

# Annotate every bar with its value
for i, (idx, row) in enumerate(df_bar.iterrows()):
    ax.annotate(f"{row['Japan']}",
                xy=(i, row['Japan']),
                ha='center', va='bottom',
                fontsize=9, fontweight='bold', color='darkred')

plt.tight_layout()
plt.show()
````

````python
# ── Horizontal bar chart ──

ax = df_continents.plot(
    kind='barh',
    figsize=(12, 6),
    color=['#e6194B', '#f58231', '#ffe119', '#3cb44b', '#4363d8', '#911eb4'],
    edgecolor='black',
    linewidth=1,
    alpha=0.85,
    fontsize=12,
    grid=True,
    legend=False,
)

ax.set_title('Immigration to Canada by Continent',
             fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('Total Immigrants', fontsize=13)
ax.set_ylabel('Continent', fontsize=13)

# Annotate bar ends with values
for i, val in enumerate(df_continents['Total']):
    ax.text(val + 2000, i, f'{val:,}', va='center', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
````

````python
# ── Grouped (multi-series) bar chart ──

df_grouped = df_top5[['India', 'China', 'UK']].tail(10)

ax = df_grouped.plot(
    kind='bar',
    figsize=(14, 6),
    color=['#e6194B', '#4363d8', '#3cb44b'],
    edgecolor='white',
    linewidth=0.8,
    width=0.8,
    alpha=0.85,
    fontsize=11,
    grid=True,
    rot=0,
)

ax.set_title('India vs China vs UK — Last 10 Years',
             fontsize=16, fontweight='bold', pad=15)
ax.set_ylabel('Number of Immigrants', fontsize=13)
ax.set_xlabel('Year', fontsize=13)
ax.legend(fontsize=12, frameon=True, shadow=True, loc='upper right')

plt.tight_layout()
plt.show()
````

## 4. Pie Charts

Pie charts display **numeric proportions** by dividing a circle into proportional slices. Best used when showing parts of a whole.

**All Parameters:**
| Parameter | Description |
|-----------|-------------|
| `kind='pie'` | Specifies pie chart |
| `autopct='%1.1f%%'` | Format string for percentage labels |
| `startangle` | Rotation in degrees counterclockwise |
| `shadow` | 3D depth effect |
| `labels` | Wedge labels (set `None` to hide) |
| `pctdistance` | Distance of % label from centre (default 0.6) |
| `labeldistance` | Distance of label from centre (default 1.1) |
| `colors` | Custom colour list for slices |
| `explode` | Tuple offsetting specific wedges |
| `wedgeprops` | Dict of wedge properties (e.g. `{'edgecolor': 'black', 'linewidth': 1}`) |
| `textprops` | Dict of text properties (e.g. `{'fontsize': 12}`) |
| `radius` | Pie radius (default 1) |
| `counterclock` | Draw counterclockwise (default True) |
| `frame` | Show axes frame |

````python
# ── Basic pie chart with full customisation ──

colours = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']

ax = df_continents['Total'].plot(
    kind='pie',
    figsize=(8, 8),
    autopct='%1.1f%%',                  # show percentages to 1 decimal
    startangle=90,                      # start from 12 o'clock
    shadow=True,                        # 3D shadow effect
    colors=colours,                     # custom slice colours
    explode=(0.05, 0.05, 0.05, 0.05, 0.05, 0.05),  # slight explode on all
    pctdistance=0.75,                   # % label position
    labeldistance=1.15,                 # category label position
    wedgeprops={'edgecolor': 'black',   # black border on each wedge
                'linewidth': 1.2},
    textprops={'fontsize': 12},         # text size
    counterclock=False,                 # clockwise direction
)

ax.set_title('Immigration to Canada by Continent [1980–2013]',
             fontsize=15, fontweight='bold', pad=20)
ax.set_ylabel('')  # hide default ylabel
plt.axis('equal')  # ensures circular shape

plt.tight_layout()
plt.show()
````

````python
# ── Pie chart with exploded wedges, external legend, hidden labels ──

explode_list = [0.1, 0, 0, 0, 0.1, 0.1]   # emphasise specific slices
colours = ['#e6194B', '#f58231', '#ffe119', '#3cb44b', '#4363d8', '#911eb4']

ax = df_continents['Total'].plot(
    kind='pie',
    figsize=(10, 8),
    autopct='%1.1f%%',
    startangle=140,
    shadow=True,
    labels=None,                            # hide on-chart labels
    pctdistance=0.85,
    colors=colours,
    explode=explode_list,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2},
    textprops={'fontsize': 13, 'fontweight': 'bold', 'color': 'white'},
)

ax.set_title('Immigration by Continent — Exploded View',
             fontsize=15, fontweight='bold', pad=20)
ax.set_ylabel('')
plt.axis('equal')

# External legend
plt.legend(labels=df_continents.index, loc='center left',
           bbox_to_anchor=(1, 0.5), fontsize=12,
           frameon=True, shadow=True, title='Continent',
           title_fontsize=13)

plt.tight_layout()
plt.show()
````

## 5. Box Plots

Box plots represent the **distribution of data** through five key dimensions:
1. **Minimum** — Smallest value (excluding outliers)
2. **First Quartile (Q1)** — 25th percentile
3. **Median (Q2)** — 50th percentile
4. **Third Quartile (Q3)** — 75th percentile
5. **Maximum** — Largest value (excluding outliers)

Points beyond **Q1 − 1.5×IQR** and **Q3 + 1.5×IQR** are plotted as **outliers**.

**All Parameters:**
| Parameter | Description |
|-----------|-------------|
| `kind='box'` | Specifies box plot |
| `vert=True/False` | Vertical (default) or horizontal |
| `color` | Dict with keys: `boxes`, `whiskers`, `medians`, `caps`, `fliers` |
| `patch_artist=True` | Fill boxes with colour (required for coloured boxes) |
| `widths` | Box width |
| `notch=True` | Notched box plot (confidence interval around median) |
| `showmeans=True` | Show mean marker |
| `meanline=True` | Show mean as line instead of marker |
| `showfliers=True/False` | Show/hide outlier points |
| `flierprops` | Dict for outlier marker style |
| `medianprops` | Dict for median line style |
| `meanprops` | Dict for mean marker style |
| `boxprops` | Dict for box style |
| `whiskerprops` | Dict for whisker line style |
| `capprops` | Dict for cap (T-bar) style |

> **Tip:** Use `df.describe()` to view quartile percentages and statistical summaries.

````python
# ── Basic box plot ──

ax = df_japan.plot(
    kind='box',
    figsize=(8, 6),
    fontsize=12,
    grid=True,
)

ax.set_title('Box Plot of Japanese Immigrants (1980–2013)',
             fontsize=15, fontweight='bold', pad=15)
ax.set_ylabel('Number of Immigrants', fontsize=13)

plt.tight_layout()
plt.show()
````

````python
# ── Highly customised box plot using matplotlib directly ──

fig, ax = plt.subplots(figsize=(12, 6))

bp = ax.boxplot(
    [df_top5['India'], df_top5['China'], df_top5['UK'],
     df_top5['Philippines'], df_top5['Pakistan']],
    labels=['India', 'China', 'UK', 'Philippines', 'Pakistan'],
    patch_artist=True,                   # fill boxes with colour
    notch=True,                          # notched boxes
    widths=0.6,                          # box width
    showmeans=True,                      # show mean marker
    meanline=False,                      # mean as point, not line
    showfliers=True,                     # show outliers
    flierprops=dict(marker='D',          # outlier: diamond shape
                    markerfacecolor='red',
                    markersize=7,
                    markeredgecolor='black',
                    alpha=0.7),
    medianprops=dict(color='black',      # median line: black, thick
                     linewidth=2.5),
    meanprops=dict(marker='^',           # mean: triangle marker
                   markerfacecolor='gold',
                   markersize=10,
                   markeredgecolor='black'),
    whiskerprops=dict(color='grey',      # whisker lines: grey, dashed
                      linewidth=1.5,
                      linestyle='--'),
    capprops=dict(color='grey',          # cap lines: grey
                  linewidth=2),
)

# Colour each box individually
box_colours = ['#e6194B', '#4363d8', '#3cb44b', '#f58231', '#911eb4']
for patch, colour in zip(bp['boxes'], box_colours):
    patch.set_facecolor(colour)
    patch.set_alpha(0.7)
    patch.set_edgecolor('black')
    patch.set_linewidth(1.5)

ax.set_title('Immigration Distribution by Country (1980–2013)',
             fontsize=16, fontweight='bold', pad=15)
ax.set_ylabel('Number of Immigrants', fontsize=13)
ax.set_xlabel('Country', fontsize=13)
ax.tick_params(axis='both', labelsize=12)
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_facecolor('#f9f9f9')              # light grey background

plt.tight_layout()
plt.show()
````

````python
# ── Horizontal box plot ──

ax = df_top5.plot(
    kind='box',
    vert=False,                # horizontal orientation
    figsize=(12, 6),
    fontsize=12,
    grid=True,
    color=dict(
        boxes='steelblue',     # box face
        whiskers='grey',       # whisker lines
        medians='red',         # median line
        caps='black',          # cap lines
    ),
)

ax.set_title('Horizontal Box Plot — Top 5 Countries',
             fontsize=15, fontweight='bold', pad=15)
ax.set_xlabel('Number of Immigrants', fontsize=13)

plt.tight_layout()
plt.show()
````

## 6. Scatter Plots

Scatter plots map **independent and dependent variables** on a 2D graph without connecting datapoints. Useful for identifying correlations and trends.

**All Parameters:**
| Parameter | Description |
|-----------|-------------|
| `kind='scatter'` | Specifies scatter plot |
| `x` | Column name for x-axis |
| `y` | Column name for y-axis |
| `c` | Colour — column name for colour-mapping, or fixed colour |
| `colormap` / `cmap` | Colourmap when `c` is numeric |
| `colorbar` | Show colourbar (when using cmap) |
| `s` | Marker size (scalar or array) |
| `marker` | Marker shape (`'o'`, `'s'`, `'^'`, `'D'`, `'*'`, `'+'`, `'x'`, etc.) |
| `alpha` | Transparency |
| `edgecolors` | Marker edge colour |
| `linewidths` | Marker edge width |

````python
# ── Basic scatter plot with full customisation ──

ax = df_tot.plot(
    kind='scatter',
    x='year',
    y='total',
    figsize=(12, 6),
    color='darkblue',                   # point colour
    s=80,                               # marker size
    marker='o',                         # circle marker
    alpha=0.7,
    edgecolors='black',                 # marker border
    linewidths=0.8,                     # border width
    fontsize=12,
    grid=True,
)

ax.set_title('Total Immigration to Canada (1980–2013)',
             fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('Year', fontsize=13)
ax.set_ylabel('Cumulative Immigrants', fontsize=13)

plt.tight_layout()
plt.show()
````

````python
# ── Scatter plot with regression line ──

x = df_tot['year']
y = df_tot['total']
fit = np.polyfit(x, y, deg=1)          # linear regression (degree 1)

ax = df_tot.plot(
    kind='scatter',
    x='year', y='total',
    figsize=(12, 6),
    color='#4363d8',
    s=90,
    marker='s',                         # square markers
    alpha=0.7,
    edgecolors='black',
    linewidths=0.8,
    grid=True,
    fontsize=12,
)

# Overlay regression line
ax.plot(x, fit[0] * x + fit[1],
        color='red', linewidth=2.5, linestyle='--',
        label=f'y = {fit[0]:.1f}x + {fit[1]:.0f}')

ax.set_title('Total Immigration with Regression Line',
             fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('Year', fontsize=13)
ax.set_ylabel('Cumulative Immigrants', fontsize=13)
ax.legend(fontsize=13, loc='upper left', frameon=True, shadow=True)

print(f'Slope (immigrants/year): {fit[0]:,.1f}')
print(f'Y-intercept: {fit[1]:,.0f}')

plt.tight_layout()
plt.show()
````

````python
# ── Scatter plot with colour-mapped third variable ──

ax = df_tot.plot(
    kind='scatter',
    x='year', y='total',
    c='total',                          # colour mapped to 'total' column
    colormap='viridis',                 # colourmap name
    colorbar=True,                      # show colourbar legend
    s=120,
    marker='D',                         # diamond marker
    edgecolors='black',
    linewidths=0.8,
    alpha=0.85,
    figsize=(12, 6),
    fontsize=12,
    grid=True,
)

ax.set_title('Immigration — Colour Mapped by Total',
             fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('Year', fontsize=13)
ax.set_ylabel('Cumulative Immigrants', fontsize=13)

plt.tight_layout()
plt.show()
````

## 7. Bubble Plots

Bubble plots extend scatter plots to **three dimensions** by using the **size** (`s` parameter) to represent a third variable (weight).

**Normalisation formula** to scale values to [0, 1] range:

$$X' = \frac{X - X_{min}}{X_{max} - X_{min}}$$

**Key additions over scatter:**
- `s` — Array of normalised values scaled up (e.g. `norm_values * 2000 + 10`)
- `alpha` — Transparency is critical for overlapping bubbles
- Overlay multiple series on the same axes using `ax=` parameter

````python
# ── Bubble plot with two countries overlaid ──

# Normalise values to [0, 1] for sizing
norm_brazil = (df_bubble['Brazil'] - df_bubble['Brazil'].min()) / \
              (df_bubble['Brazil'].max() - df_bubble['Brazil'].min())

norm_argentina = (df_bubble['Argentina'] - df_bubble['Argentina'].min()) / \
                 (df_bubble['Argentina'].max() - df_bubble['Argentina'].min())

# First scatter: Brazil
ax = df_bubble.plot(
    kind='scatter',
    x='Year', y='Brazil',
    figsize=(14, 8),
    alpha=0.5,
    color='green',
    s=norm_brazil * 2000 + 10,          # sized by normalised values
    marker='o',
    edgecolors='darkgreen',
    linewidths=1.2,
    label='Brazil',
    xlim=(1978, 2016),
    grid=True,
    fontsize=12,
)

# Second scatter: Argentina (overlaid on same axes)
df_bubble.plot(
    kind='scatter',
    x='Year', y='Argentina',
    alpha=0.5,
    color='blue',
    s=norm_argentina * 2000 + 10,
    marker='s',                          # square markers for contrast
    edgecolors='darkblue',
    linewidths=1.2,
    label='Argentina',
    ax=ax,                               # overlay on same axes
)

ax.set_title('Immigration from Brazil & Argentina (1980–2013)\nBubble Size = Normalised Immigration',
             fontsize=15, fontweight='bold', pad=15)
ax.set_ylabel('Number of Immigrants', fontsize=13)
ax.set_xlabel('Year', fontsize=13)
ax.legend(loc='upper left', fontsize=13, frameon=True,
          shadow=True, markerscale=0.5)

plt.tight_layout()
plt.show()
````

## Subplots

Combine multiple plots in one figure.

````python
# ── Method 1: fig.add_subplot() ──

fig = plt.figure(figsize=(16, 6))

ax0 = fig.add_subplot(1, 2, 1)       # 1 row, 2 cols, position 1
ax1 = fig.add_subplot(1, 2, 2)       # 1 row, 2 cols, position 2

df_japan.plot(kind='box', ax=ax0, grid=True,
              color=dict(boxes='steelblue', medians='red',
                         whiskers='grey', caps='black'))
ax0.set_title('Box Plot — Japan', fontsize=14, fontweight='bold')
ax0.set_ylabel('Immigrants', fontsize=12)

df_japan.plot(kind='line', ax=ax1, grid=True,
              color='steelblue', linewidth=2,
              marker='o', markersize=4, markerfacecolor='red')
ax1.set_title('Line Plot — Japan', fontsize=14, fontweight='bold')
ax1.set_ylabel('Immigrants', fontsize=12)
ax1.set_xlabel('Year', fontsize=12)
ax1.legend().set_visible(False)

plt.tight_layout()
plt.show()
````

## Data Analysis with Pandas

```python
import pandas as pd

# Basic exploration
df.shape  # (rows, columns)
df.info()  # Data types, non-null counts
df.describe()  # Summary statistics for numeric columns

# Single column analysis
df['Q1'].value_counts()  # Frequency table
df['Q1'].value_counts(normalize=True)  # Percentages
df['age'].mean()  # Average
df['age'].median()  # Median

# Cross-tabulation (relationship between 2 variables)
pd.crosstab(df['gender'], df['party_preference'])
pd.crosstab(df['gender'], df['party_preference'], normalize='index')  # Row %

# Filter and analyze
young_voters = df[df['age'] < 30]
young_voters['party_preference'].value_counts()

# Group analysis
df.groupby('region')['satisfaction'].mean()
df.groupby(['region', 'gender'])['satisfaction'].agg(['mean', 'count', 'std'])

# Correlation between numeric variables
df[['age', 'income', 'satisfaction']].corr()
```

## Data Analysis with Pyreadstat

```python
import pyreadstat

df, meta = pyreadstat.read_sav('your_file.sav')

# The KEY advantage: value labels (e.g., 1='Male', 2='Female')
print(meta.variable_value_labels)

# Example: Q1 might have codes like 1, 2, 3
# But they mean 'Strongly Agree', 'Agree', 'Disagree'

# Get value labels for specific variable
q1_labels = meta.variable_value_labels.get('Q1', {})
print(q1_labels)  # {1: 'Strongly Agree', 2: 'Agree', ...}

# Replace codes with labels
df['Q1_labeled'] = df['Q1'].map(q1_labels)
df['Q1_labeled'].value_counts()  # Now shows actual meanings!

# Or apply to entire dataframe
for col in df.columns:
    if col in meta.variable_value_labels:
        df[f'{col}_labeled'] = df[col].map(meta.variable_value_labels[col])

# Get question text
question_text = meta.column_names_to_labels.get('Q15')
print(f"Q15 asks: {question_text}")
```

## Combined approach ##

```python
import pandas as pd
import pyreadstat

df, meta = pyreadstat.read_sav('your_file.sav')

# Create a readable version with labels
df_readable = df.copy()
for col in df.columns:
    if col in meta.variable_value_labels:
        df_readable[col] = df[col].map(meta.variable_value_labels[col])

# Now analyze the readable version
df_readable['party_preference'].value_counts()
# Shows: "Conservative: 150, Labour: 200..." 
# Instead of: "1: 150, 2: 200..."

# Cross-tab with actual labels
pd.crosstab(df_readable['gender'], df_readable['voting_intention'])
```

# how to do a cross-tab between multiple variables (such as Otherspont)

```python
import pandas as pd

# ── STEP 1: Create derived demographic columns ────────────────────────────────

# Region: Yerevan vs Other (based on numeric S2 in df)
# Replace 1 with whatever code represents Yerevan in your S2 column
df_readable['Region'] = df['S2'].apply(lambda x: 'Yerevan' if x == 1 else 'Other')

# Income: Up to 400k (D51) vs Above 400k (D52)
# 99 is treated as missing/refused
df_readable['Income'] = df.apply(
    lambda row: 'Up to 400k' if pd.notna(row['D51']) and row['D51'] != 99
                else ('Above 400k' if pd.notna(row['D52']) and row['D52'] != 99
                else None),
    axis=1
)

# ── STEP 2: Define the question columns (multiple response) ───────────────────
# These are the binary 0/1 columns representing each answer option
q_cols = [
    'B1otherSpont_1', 'B1otherSpont_3', 'B1otherSpont_4', 'B1otherSpont_5',
    'B1otherSpont_9', 'B1otherSpont_10', 'B1otherSpont_11'
]

# ── STEP 3: Map column codes to readable labels (pulled from SPSS metadata) ───
col_labels = {c: meta.column_names_to_labels.get(c, c) for c in q_cols}

# ── STEP 4: Define demographic variables to cross by ─────────────────────────
# Key = column name in df_readable, Value = label shown in output
demo_vars = {
    'S1':     'Gender',
    'Region': 'Region',
    'Age':    'Age',
    'Income': 'Income'
}

# ── STEP 5: Build working dataframe ──────────────────────────────────────────
# Numeric 0/1 from df for the math, readable labels from df_readable for display
work = df[q_cols].join(df_readable[list(demo_vars.keys())])

# ── STEP 6: Crosstab function ─────────────────────────────────────────────────
# For each demographic group, count how many respondents selected each option
def crosstab_by_demo(demo_col, demo_label):
    rows = []
    for val in work[demo_col].dropna().unique():
        subset = work[work[demo_col] == val]
        n = len(subset)
        row = {'Demographic': demo_label, 'Group': val, 'N': n}
        for col in q_cols:
            row[col_labels[col]] = subset[col].sum().astype(int)
        rows.append(row)
    return rows

# ── STEP 7: Build the full crosstab table ─────────────────────────────────────
all_rows = []
for demo_col, demo_label in demo_vars.items():
    all_rows.extend(crosstab_by_demo(demo_col, demo_label))

result = pd.DataFrame(all_rows)

# ── STEP 8: Display & export ──────────────────────────────────────────────────
print(result.to_string(index=False))

# Export to Excel (utf-8-sig ensures Armenian characters display correctly)
result.to_excel('crosstab_B1.xlsx', index=False)
print("Exported to crosstab_B1.xlsx")
```

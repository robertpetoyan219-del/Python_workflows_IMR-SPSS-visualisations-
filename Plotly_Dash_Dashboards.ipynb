# Part 1 — Plotly Express (`px`)

**Plotly Express** is the high-level wrapper for Plotly. Every `px` function returns a `plotly.graph_objects.Figure`, so you can further customize it with `.update_layout()`, `.update_traces()`, etc.

All examples below use built-in Plotly datasets so the notebook runs without external files.

## 1.1 — Scatter Plot

Use `px.scatter()` to plot individual data points. Map columns to x, y, color, size, and hover info.

**What to change for your own data:**
- Replace `px.data.iris()` with your own DataFrame
- Change `x=`, `y=` to your column names
- `color=` groups points by a categorical column (remove if not needed)
- `size=` maps a numeric column to bubble size (remove for a plain scatter)
- `hover_data=` lists extra columns shown on hover

````python
df_iris = px.data.iris()

fig = px.scatter(
    df_iris,
    x='sepal_width',        # column for the x-axis
    y='sepal_length',       # column for the y-axis
    color='species',        # color-code points by this column
    size='petal_length',    # bubble size mapped to this column
    hover_data=['petal_width'],  # extra info shown on hover
    title='Iris — Sepal Width vs. Length'
)
fig.show()
````

## 1.2 — Line Chart

Use `px.line()` for time-series or sequential data. Each unique value in `color=` becomes a separate line.

**What to change for your own data:**
- `x=` should be your time or sequence column
- `y=` is the value column you want to plot
- `color=` splits the data into separate lines (e.g., by country, category)
- Add `markers=True` if you want dots at each data point

````python
df_life = px.data.gapminder().query("country in ['Canada', 'United Kingdom', 'Australia']")

fig = px.line(
    df_life,
    x='year',               # time axis
    y='lifeExp',            # values to plot
    color='country',        # one line per country
    markers=True,           # show data-point dots
    title='Life Expectancy Over Time'
)
fig.show()
````

## 1.3 — Bar Chart

Use `px.bar()` for comparing quantities across categories.

**What to change for your own data:**
- `x=` is the categorical column (labels on the x-axis)
- `y=` is the numeric column (bar height)
- `color=` stacks or groups bars by another category
- Set `barmode='group'` in `update_layout()` to show grouped bars instead of stacked

````python
df_tips = px.data.tips()

fig = px.bar(
    df_tips,
    x='day',                # categories on x-axis
    y='total_bill',         # bar height
    color='sex',            # stack by gender
    title='Total Bill by Day'
)
# Switch from stacked to grouped bars:
# fig.update_layout(barmode='group')
fig.show()
````

## 1.4 — Histogram

Use `px.histogram()` to visualize the distribution of a single variable.

**What to change for your own data:**
- `x=` is the numeric column whose distribution you want to see
- `nbins=` sets the number of bins (omit to let Plotly choose automatically)
- `color=` overlays separate distributions for different groups
- `marginal=` adds a subplot above/beside — options: `'rug'`, `'box'`, `'violin'`

````python
fig = px.histogram(
    df_tips,
    x='total_bill',         # column to bin
    nbins=20,               # number of bins
    color='sex',            # overlay distributions by group
    marginal='box',         # marginal subplot: 'rug', 'box', 'violin'
    title='Distribution of Total Bills'
)
fig.show()
````

## 1.5 — Pie Chart

Use `px.pie()` to show part-of-whole relationships.

**What to change for your own data:**
- `names=` is the categorical column (slice labels)
- `values=` is the numeric column (slice size)
- `hole=` adds a donut hole — set between 0 and 1 (e.g., `0.4` for a donut chart)

````python
fig = px.pie(
    df_tips,
    names='day',            # slice labels
    values='total_bill',    # slice sizes
    title='Share of Total Bill by Day',
    hole=0.3                # donut style (0 = full pie)
)
fig.show()
````

## 1.6 — Bubble Chart

A bubble chart is a scatter plot where `size=` maps a third variable to the point area.

**What to change for your own data:**
- Same parameters as `px.scatter()`, but `size=` is essential
- `size_max=` caps the largest bubble size in pixels
- `animation_frame=` and `animation_group=` add animated transitions (great for time data)

````python
df_gap = px.data.gapminder().query('year == 2007')

fig = px.scatter(
    df_gap,
    x='gdpPercap',
    y='lifeExp',
    size='pop',             # bubble size = population
    size_max=60,            # max bubble radius in pixels
    color='continent',
    hover_name='country',   # bold country name at top of hover
    log_x=True,             # logarithmic x-axis
    title='GDP vs Life Expectancy (2007)'
)
fig.show()
````

## 1.7 — Sunburst Chart

Use `px.sunburst()` to show hierarchical part-of-whole data in concentric rings.

**What to change for your own data:**
- `path=` is a list of columns defining the hierarchy — outermost ring first (e.g., `['continent', 'country']`)
- `values=` is the numeric column that determines slice sizes
- `color=` colors the slices by a numeric or categorical column

````python
df_gap_2007 = px.data.gapminder().query('year == 2007')

fig = px.sunburst(
    df_gap_2007,
    path=['continent', 'country'],   # hierarchy levels (outer → inner)
    values='pop',                    # slice sizes
    color='lifeExp',                 # color by life expectancy
    color_continuous_scale='RdBu',   # diverging color scale
    title='World Population Sunburst (2007)'
)
fig.show()
````

# Part 2 — Plotly Graph Objects (`go`)

**Graph Objects** is the low-level interface. Use it when you need:
- Multiple trace types on the same axes (e.g., bar + line)
- Fine-grained control over every visual property
- Subplots with `make_subplots()`

The basic pattern is always:
```python
fig = go.Figure()                          # create empty figure
fig.add_trace(go.Scatter(...))             # add one or more traces
fig.update_layout(title='My Chart')        # customise the layout
fig.show()                                 # display
```

## 2.1 — Scatter / Line with `go.Scatter`

`go.Scatter` handles both scatter plots and line charts — controlled by `mode`.

**What to change for your own data:**
- Replace the `x=` and `y=` arrays with your own data (lists, NumPy arrays, or DataFrame columns)
- `mode=` controls appearance: `'markers'` (scatter), `'lines'` (line), `'lines+markers'` (both)
- `name=` sets the legend label for this trace

````python
# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales_a = [150, 200, 170, 220, 250, 300]
sales_b = [120, 180, 160, 190, 210, 260]

fig = go.Figure()

# Trace 1: line + markers
fig.add_trace(go.Scatter(
    x=months,
    y=sales_a,
    mode='lines+markers',          # 'markers', 'lines', or 'lines+markers'
    name='Product A',              # legend label
    marker=dict(size=10, color='royalblue')
))

# Trace 2: markers only
fig.add_trace(go.Scatter(
    x=months,
    y=sales_b,
    mode='markers',
    name='Product B',
    marker=dict(size=10, symbol='diamond', color='tomato')
))

fig.update_layout(title='Monthly Sales', xaxis_title='Month', yaxis_title='Sales ($)')
fig.show()
````

## 2.2 — Bar Chart with `go.Bar`

**What to change for your own data:**
- `x=` is the category list; `y=` is the value list
- Add multiple `go.Bar` traces for grouped/stacked bars
- Use `fig.update_layout(barmode='group')` for side-by-side, or `'stack'` for stacked

````python
categories = ['Q1', 'Q2', 'Q3', 'Q4']

fig = go.Figure()
fig.add_trace(go.Bar(x=categories, y=[300, 450, 500, 620], name='2023'))
fig.add_trace(go.Bar(x=categories, y=[350, 420, 530, 700], name='2024'))

fig.update_layout(
    title='Quarterly Revenue',
    barmode='group',       # 'group' = side-by-side, 'stack' = stacked
    xaxis_title='Quarter',
    yaxis_title='Revenue ($k)'
)
fig.show()
````

## 2.3 — Combining Different Trace Types

One of the main advantages of `go` over `px` is mixing trace types on the same figure.

````python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue = [120, 150, 170, 160, 200, 220]
target  = [130, 140, 160, 170, 180, 200]

fig = go.Figure()

# Bars for actual revenue
fig.add_trace(go.Bar(x=months, y=revenue, name='Revenue', marker_color='steelblue'))

# Line for target
fig.add_trace(go.Scatter(
    x=months, y=target, mode='lines+markers',
    name='Target', line=dict(color='red', dash='dash')
))

fig.update_layout(title='Revenue vs Target')
fig.show()
````

## 2.4 — Subplots with `make_subplots()`

Use `make_subplots()` to place multiple charts in a grid. Each trace targets a specific cell with `row=` and `col=`.

**What to change for your own data:**
- `rows=` and `cols=` control the grid size
- `subplot_titles=` sets the title above each cell
- When adding traces, specify `row=` and `col=` to position them

````python
from plotly.subplots import make_subplots

fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=('Sales Trend', 'Sales Distribution')
)

# Left panel: line chart
fig.add_trace(
    go.Scatter(x=months, y=sales_a, mode='lines+markers', name='Product A'),
    row=1, col=1
)

# Right panel: histogram
fig.add_trace(
    go.Histogram(x=sales_a + sales_b, name='All Sales', nbinsx=8),
    row=1, col=2
)

fig.update_layout(title_text='Side-by-Side Subplots', showlegend=True)
fig.show()
````

## 3.5 — Callbacks — Making It Interactive

Callbacks are the core of Dash interactivity. A callback is a Python function that **automatically runs** whenever a user changes an input component (dropdown, slider, etc.).

### Anatomy of a callback
```python
from dash import callback, Input, Output

@callback(
    Output('graph-id', 'figure'),      # what to update (component id, property)
    Input('dropdown-id', 'value')      # what triggers the update
)
def update_graph(selected_value):       # function receives the Input value
    fig = px.scatter(df, x='col1', y=selected_value)
    return fig                          # returned value goes to the Output
```

**Key rules:**
1. The number of function arguments must match the number of `Input()`s
2. The function must return a value for every `Output()`
3. `Input('id', 'property')` — first arg is the component `id`, second is the property to watch
4. `Output('id', 'property')` — first arg is the component `id`, second is the property to update

# Part 4 — Cheat Sheets & Quick Reference

## 4.1 — Dash App Skeleton

Every Dash app follows this structure:

```python
from dash import Dash, html, dcc, callback, Input, Output
import plotly.express as px

app = Dash()                          # 1. Create the app

app.layout = html.Div([              # 2. Define the layout
    dcc.Dropdown(id='my-input', options=[...], value='...'),
    dcc.Graph(id='my-output')
])

@callback(                            # 3. Define callbacks
    Output('my-output', 'figure'),
    Input('my-input', 'value')
)
def update(val):
    return px.scatter(df, x='col', y=val)

if __name__ == '__main__':            # 4. Run the server
    app.run(debug=True)
```

## 4.2 — Callback Patterns Summary

| Pattern | Decorator |
|---|---|
| Single input → single output | `@callback(Output('id', 'prop'), Input('id', 'prop'))` |
| Multiple inputs → single output | `@callback(Output(...), Input(...), Input(...))` |
| Single input → multiple outputs | `@callback(Output(...), Output(...), Input(...))` |
| Multiple inputs → multiple outputs | `@callback(Output(...), Output(...), Input(...), Input(...))` |

**Common callback mistakes:**
- Using `@app_callback` instead of `@app.callback` (or `@callback` with the modern import)
- Mismatched `id` strings between layout and callback
- Wrong number of `return` values vs. number of `Output`s
- Forgetting that function arguments are positional — they match `Input`s in order

## 4.3 — Dropdown Options Formats

Three ways to specify `dcc.Dropdown` options:

```python
# 1. Simple list — label and value are the same
dcc.Dropdown(options=['NYC', 'MTL', 'SF'], value='NYC')

# 2. List of dicts — label shown to user, value used in callback
dcc.Dropdown(options=[
    {'label': 'New York City', 'value': 'NYC'},
    {'label': 'Montreal',      'value': 'MTL'},
    {'label': 'San Francisco', 'value': 'SF'}
], value='NYC')

# 3. Multi-select — user can pick more than one
dcc.Dropdown(options=[...], value=['NYC'], multi=True)
```

## 4.4 — Styling Tips

Dash uses inline `style=` dictionaries with **camelCase** property names (not CSS kebab-case).

| CSS property | Dash style key |
|---|---|
| `text-align` | `textAlign` |
| `font-size` | `fontSize` |
| `background-color` | `backgroundColor` |
| `margin-top` | `marginTop` |
| `border-radius` | `borderRadius` |

**Side-by-side layout:**
```python
html.Div([
    html.Div([...], style={'width': '49%', 'display': 'inline-block'}),
    html.Div([...], style={'width': '49%', 'display': 'inline-block', 'float': 'right'})
])
```

# Part 1 — Word Clouds

A **word cloud** is a visual representation of text data where the size of each word indicates its frequency or importance.

## 1.1 — Basic Word Cloud from a String

The simplest approach: pass a string of text to `WordCloud().generate()`.

```python
text = """
Python is a great programming language for data science and machine learning.
Data visualization helps us understand data patterns. Python provides many
libraries for data analysis such as pandas numpy and matplotlib. Machine learning
models can be built using scikit-learn and tensorflow. Data science combines
statistics programming and domain knowledge. Python is widely used in data
science because of its simplicity and powerful libraries. Data analysis and
data visualization are key skills for any data scientist. Python makes it
easy to work with data and build machine learning models.
"""

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    max_words=100
).generate(text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Basic Word Cloud from Text')
plt.show()
```

## 1.2 — Customizing Stopwords

**Stopwords** are common words ("the", "is", "and", etc.) that are excluded from the cloud. The library includes a default set, and you can add your own.

```python
# Start with the built-in stopwords and add custom ones
stopwords = set(STOPWORDS)
stopwords.update(['us', 'using', 'used', 'many', 'provides', 'makes'])

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    stopwords=stopwords,
    colormap='viridis',
    max_words=80
).generate(text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud with Custom Stopwords & Colormap')
plt.show()
```

## 1.3 — Word Cloud from a Frequency Dictionary

Instead of raw text, you can pass a `{word: count}` dictionary using `generate_from_frequencies()`.

```python
word_freq = {
    'Python': 50, 'Data': 45, 'Science': 35, 'Machine Learning': 30,
    'Visualization': 28, 'Analysis': 25, 'Statistics': 20, 'Pandas': 18,
    'NumPy': 17, 'Matplotlib': 16, 'Deep Learning': 14, 'TensorFlow': 12,
    'Neural Network': 11, 'Algorithm': 10, 'Regression': 9, 'Classification': 8,
    'Clustering': 7, 'SQL': 15, 'Database': 6, 'API': 5
}

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='black',
    colormap='Set2',
    max_words=50
).generate_from_frequencies(word_freq)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud from Frequency Dictionary', color='white')
plt.show()
```

## 1.4 — Word Cloud from a DataFrame Column

A common real-world use case: generate a cloud from a column of text data in a Pandas DataFrame.

```python
# Sample DataFrame with feedback text
df = pd.DataFrame({
    'feedback': [
        'Great course very informative and helpful',
        'The instructor was excellent and engaging',
        'Loved the hands-on labs and projects',
        'Python programming was well explained',
        'Data visualization section was the best part',
        'Helpful course for learning data science',
        'Great introduction to machine learning concepts',
        'The projects helped me understand real world applications',
        'Excellent content on data analysis with Python',
        'Very practical and easy to follow along'
    ]
})

# Join all rows into a single string
all_text = ' '.join(df['feedback'])

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    colormap='plasma',
    min_font_size=10
).generate(all_text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud from DataFrame Column')
plt.show()
```

## 1.6 — WordCloud Key Parameters Reference

| Parameter | Default | Description |
|---|---|---|
| `width` | 400 | Width of the canvas (pixels) |
| `height` | 200 | Height of the canvas (pixels) |
| `max_words` | 200 | Maximum number of words displayed |
| `min_font_size` | 4 | Smallest font size used |
| `max_font_size` | None | Largest font size (auto if None) |
| `background_color` | `'black'` | Background color of the image |
| `colormap` | `'viridis'` | Matplotlib colormap for word colors |
| `stopwords` | built-in set | Words to exclude |
| `collocations` | True | Whether to include two-word phrases |
| `mask` | None | NumPy array to shape the cloud |
| `relative_scaling` | `'auto'` | How much to scale word sizes by frequency |

**Key methods:**
- `generate(text)` — generate from raw text
- `generate_from_frequencies(dict)` — generate from a word→count dictionary
- `to_file(path)` — save the cloud as an image file

# Part 2 — Geospatial Data with Folium

**Folium** builds interactive [Leaflet.js](https://leafletjs.com/) maps right inside Jupyter.  
It is completely free with no API-key limits.

## 2.1 — Creating a Basic Map

`folium.Map()` creates an interactive map. Set the center with `location=[lat, lng]` and initial zoom with `zoom_start`.

```python
# Default world map
world_map = folium.Map()
world_map
```

```python
# Map centered on Canada with zoom
canada_map = folium.Map(location=[56.130, -106.35], zoom_start=4)
canada_map
```

## 2.2 — Map Tile Styles

Change the visual style using the `tiles` parameter.

| Tile | Look |
|---|---|
| `'OpenStreetMap'` | Default street map |
| `'CartoDB positron'` | Light, minimal style |
| `'CartoDB dark_matter'` | Dark theme |

```python
# Dark-themed map
dark_map = folium.Map(
    location=[40.7128, -74.0060],  # New York City
    zoom_start=11,
    tiles='CartoDB dark_matter'
)
dark_map
```

## 2.3 — Adding Markers

Use `folium.Marker()` to pin locations on the map. Add interactivity with `popup` (click) and `tooltip` (hover).

```python
nyc_map = folium.Map(location=[40.7128, -74.0060], zoom_start=12)

# Add markers for NYC landmarks
landmarks = [
    {'name': 'Statue of Liberty',   'coords': [40.6892, -74.0445]},
    {'name': 'Empire State Building','coords': [40.7484, -73.9857]},
    {'name': 'Central Park',         'coords': [40.7829, -73.9654]},
    {'name': 'Times Square',         'coords': [40.7580, -73.9855]},
]

for lm in landmarks:
    folium.Marker(
        location=lm['coords'],
        popup=lm['name'],
        tooltip=f"Hover: {lm['name']}"
    ).add_to(nyc_map)

nyc_map
```

## 2.4 — Circle Markers

`folium.CircleMarker()` draws circles that can encode data through color, radius, and opacity.

```python
sf_map = folium.Map(location=[37.77, -122.42], zoom_start=12)

# Sample incident locations in San Francisco
incidents = [
    [37.7749, -122.4194], [37.7849, -122.4094],
    [37.7649, -122.4294], [37.7799, -122.4150],
    [37.7700, -122.4350], [37.7880, -122.3990],
    [37.7620, -122.4250], [37.7770, -122.4100],
]

for lat, lng in incidents:
    folium.CircleMarker(
        location=[lat, lng],
        radius=8,
        popup='Incident',
        color='red',
        fill=True,
        fillColor='yellow',
        fillOpacity=0.6
    ).add_to(sf_map)

sf_map
```

## 2.5 — Choropleth Maps

A **choropleth** map shades geographic regions by a data variable (e.g., population, GDP).  
You need two things:
1. **GeoJSON** — boundary coordinates for each region
2. **Data** — a DataFrame with the values to visualize

The `key_on` parameter links the GeoJSON feature property to your DataFrame column.

```python
# GeoJSON for US states (public dataset)
us_states_geo = 'https://raw.githubusercontent.com/python-visualization/folium/main/tests/us-states.json'

# Sample state-level data
state_data = pd.DataFrame({
    'State': ['AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA',
              'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD',
              'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ',
              'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC',
              'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY'],
    'Value': np.random.randint(100, 1000, size=50)
})

us_map = folium.Map(location=[48, -102], zoom_start=3)

folium.Choropleth(
    geo_data=us_states_geo,
    data=state_data,
    columns=['State', 'Value'],
    key_on='feature.id',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Random Value by State'
).add_to(us_map)

us_map
```

## 2.6 — Choropleth Color Scales

The `fill_color` parameter accepts any [Brewer color scheme](https://colorbrewer2.org/):

| Scheme | Colors |
|---|---|
| `'YlOrRd'` | Yellow → Orange → Red |
| `'YlGnBu'` | Yellow → Green → Blue |
| `'BuPu'` | Blue → Purple |
| `'PuRd'` | Purple → Red |
| `'GnBu'` | Green → Blue |
| `'OrRd'` | Orange → Red |
| `'RdYlGn'` | Red → Yellow → Green (diverging) |

## 2.7 — GeoJSON Layers with Styling

For more control over how regions are drawn, use `folium.GeoJson()` with a custom `style_function`.

```python
styled_map = folium.Map(location=[48, -102], zoom_start=3)

folium.GeoJson(
    us_states_geo,
    style_function=lambda feature: {
        'fillColor': 'green',
        'color': 'black',
        'weight': 1,
        'fillOpacity': 0.3
    },
    tooltip=folium.GeoJsonTooltip(fields=['name'], aliases=['State:'])
).add_to(styled_map)

styled_map
```

## 2.8 — Layer Control

Add a toggle so users can switch layers on and off.

```python
layered_map = folium.Map(location=[40.7128, -74.0060], zoom_start=11)

# Layer 1: Markers
marker_layer = folium.FeatureGroup(name='Landmarks')
for lm in landmarks:
    folium.Marker(location=lm['coords'], popup=lm['name']).add_to(marker_layer)
marker_layer.add_to(layered_map)

# Layer 2: Circle markers
circle_layer = folium.FeatureGroup(name='Circles')
for lm in landmarks:
    folium.CircleMarker(
        location=lm['coords'], radius=20,
        color='purple', fill=True, fillOpacity=0.3
    ).add_to(circle_layer)
circle_layer.add_to(layered_map)

# Add toggle control
folium.LayerControl().add_to(layered_map)

layered_map
```

## 2.9 — Folium Key Functions Reference

| Function | Description |
|---|---|
| `folium.Map(location, zoom_start, tiles)` | Create an interactive map |
| `folium.Marker(location, popup, tooltip, icon)` | Add a pin marker |
| `folium.CircleMarker(location, radius, color, fill, fillColor, fillOpacity)` | Add a circle marker |
| `folium.Choropleth(geo_data, data, columns, key_on, fill_color)` | Create a choropleth map |
| `folium.GeoJson(data, style_function, tooltip)` | Add a styled GeoJSON layer |
| `folium.GeoJsonTooltip(fields, aliases)` | Add hover tooltips to GeoJSON features |
| `folium.FeatureGroup(name)` | Group map elements into a named layer |
| `folium.LayerControl()` | Add a layer toggle control to the map |

## 1 — Drawing a Square

The most basic turtle program: move forward and turn, four times.

```python
from ColabTurtlePlus.Turtle import *

clearscreen()
setup(400, 400)
speed(5)

for _ in range(4):
    forward(100)
    right(90)
```

## 2 — Colorful Star

Draw a five-pointed star with some color.

```python
from ColabTurtlePlus.Turtle import *

clearscreen()
setup(400, 400)
speed(5)
color("gold")
pensize(3)

for _ in range(5):
    forward(150)
    right(144)
```

## 3 — Spiral

A spiral is made by gradually increasing the distance you move forward.

```python
from ColabTurtlePlus.Turtle import *

clearscreen()
setup(500, 500)
speed(10)
pensize(2)

colors = ["red", "blue", "green", "orange", "purple"]

for i in range(60):
    color(colors[i % len(colors)])
    forward(i * 3)
    right(91)
```

## 4 — Using the Turtle Class (Multiple Turtles)

You can also create turtle objects, just like in the standard library.

```python
from ColabTurtlePlus.Turtle import *

clearscreen()
setup(500, 400)

t1 = Turtle()
t1.color("blue")
t1.speed(6)

t2 = Turtle()
t2.color("red")
t2.speed(6)

# t1 draws a triangle
for _ in range(3):
    t1.forward(80)
    t1.left(120)

# t2 draws a hexagon offset to the right
t2.penup()
t2.forward(150)
t2.pendown()
for _ in range(6):
    t2.forward(50)
    t2.left(60)
```
